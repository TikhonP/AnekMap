```python
import requests


def send(asd):
    req = {
        "text_preview": asd["text"][:100],
        "text": asd["text_main"],
        "date_created": asd["date_created"],
        "x": int(asd['2dim_vec2'][0]),
        "y": int(asd['2dim_vec2'][1]),
        "likes": int(asd["likes"]),
        "views": int(asd["views"][:-1])*1000,
        "href": asd["href"],
    }
    url = "http://192.168.234.30:8000/api/v1/aneks/anek/"
    a = requests.post(url, data=req)
    if a.status_code!=201:
        return a.text
    else:
        return a.json()
```

```python
used_tags = {}
```

```python
def createTag(name, url):
    req = {
        "name": name,
        "aneks": [url],
    }
    
    url = "http://192.168.234.30:8000/api/v1/aneks/tag/"
    
    a = requests.post(url, data=req)
    
#     print(req)
    
    if a.status_code!=201:
        return a.text
    else:
        return req, a.json()
```

```python
def updateTag(req, url, newurl):
    
    req["aneks"].append(newurl)
    
#     print(req, url, newurl)
    
    a = requests.put(url, data=req)
    
    if a.status_code!=201:
        return a.text
    else:
        return req, a.json()
```

```python
import time

for n, i in enumerate(data):
    a = send(i)
    
    if type(a)==str:
        print(a)
        break
        
    for j in i['tags_main']:
        if j not in used_tags:
            req, an = createTag(j, a['url'])
            used_tags[j] = [an['url'], req]
        else:
            updateTag(used_tags[j][1], used_tags[j][0], a['url'])
            
    print(n/len(data))
    time.sleep(0.2)
```
